```python
import collections
import ast 
# select by sector 
def get_skills(filter_key,filter_value):
    filtred_df =   df[df[filter_key].str.contains(filter_value)]
    skills_2d = filtred_df['skills'].values.tolist()
    # str to list 
    skills_2d = [ast.literal_eval(x) for x in skills_2d]
    skills_1d = [item for sublist in skills_2d for item in sublist]
    # remove noise 
    skills_1d = [skill for skill in skills_1d if skill not in stop_skills]
    return skills_1d



# test 
#get_skills('location','Casablanca')
```

```python
# process skills (hard ,soft)
from skillNer.general_params import SKILL_DB

def process_skill_list(skill_lst):
    hard_skill_list = []
    soft_skill_list = []
    for skill in skill_lst : 
        # get skill type  
        skill_type = SKILL_DB[skill]['skill_type']
        #print(skill, skill_type)
        if skill_type == 'Hard Skill' : 
            # add to hard skill list 
            hard_skill_list.append(skill)
        elif skill_type == 'Soft Skill' : 
            # add to soft skill list 
            soft_skill_list.append(skill)
    
    # create counters 
    hard_skill_counter = collections.Counter(hard_skill_list)
    soft_skill_counter = collections.Counter(soft_skill_list)
    # get most common
    hard_skill_most_common = hard_skill_counter.most_common(10)
    soft_skill_most_common = soft_skill_counter.most_common(10)
    # map to skill name
    hard_skill_most_common = [(SKILL_DB[x[0]]['skill_name'],x[1]) for x in hard_skill_most_common]
    soft_skill_most_common = [(SKILL_DB[x[0]]['skill_name'],x[1]) for x in soft_skill_most_common]
    return hard_skill_most_common, soft_skill_most_common
```

```python
col_name = 'title'
filter_value = 'analyst'
skills = get_skills(col_name,filter_value)
stats = process_skill_list(skills)

import pprint 
pprint.pprint(stats)
```

```python
# get skill stats 
# skill occurence in jobs 
def get_skill_uno_stats(skill_id):
    # get skill name 
    skill_name = SKILL_DB[skill_id]['skill_name']
    # get skill count 
    skill_count = df[df['skills'].str.contains(skill_id)].shape[0]
    # get occurence dates 
    occurence_dates = df[df['skills'].str.contains(skill_id)]['date'].values.tolist()

    return skill_name, skill_count,dict(collections.Counter(occurence_dates))

# test 
get_skill_uno_stats('KS123K75YYK8VGH90NCS')
```
